```python
import pandas as pd

# Load the dataset from ChEMBL (replace 'your_dataset.csv' with the actual file path)
file_path = 'antibiotics_600.csv'
df = pd.read_csv(file_path)

# Step 1: Drop rows with missing canonical smiles
df = df.dropna(subset=['Smiles'])

# Step 2: Make the canonical smiles unique
df['canonical_smiles'] = df['Smiles'].astype(str).str.strip()  # Remove leading/trailing whitespaces
df = df.drop_duplicates(subset='Smiles', keep='first')

# Display the first few rows of the processed dataframe
print("Processed Dataset:")
print(df)

# Save the processed dataset to a new CSV file if needed
output_file_path = 'processed_600_dataset.csv'
# df.to_csv(output_file_path, index=False)
# print(f"Processed dataset saved to {output_file_path}")
```
